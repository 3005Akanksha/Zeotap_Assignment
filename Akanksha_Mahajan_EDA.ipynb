```python
# Define Numerical and Categorical Columns
numerical_features_customer = [feature for feature in customer.columns if customer[feature].dtype in ['int64', 'float64']]
categorical_features_customer = [feature for feature in customer.columns if customer[feature].dtype == 'object']

# Print the results
print("Numerical Features:", numerical_features_customer)
print("Categorical Features:", categorical_features_customer)
```

```python
#Exploring Data visualization

#Customer

fig, axs=plt.subplots(1,1, figsize=(15,7))
plt.subplot(121)
sns.histplot(data=customer,x='Region',bins=30,kde=True,color='skyblue')
plt.show()
```

```python
#product
fig, axs=plt.subplots(1,1, figsize=(15,7))
plt.subplot(121)
sns.histplot(data=product,x='Price',bins=30,kde=True,color='purple')
plt.show()
```

```python
fig, axs=plt.subplots(1,1, figsize=(15,10))
sns.histplot(data=product,x='Price',kde=True,hue='Category')
plt.show()
```

```python
# Replace less common products with 'Other'
top_n = 10
product['ProductGroup'] = product['ProductName'].apply(
    lambda x: x if x in product['ProductName'].value_counts().nlargest(top_n).index else 'Other'
)

fig, axs = plt.subplots(1, 1, figsize=(15, 7))
sns.histplot(data=product, x='Price', bins=30, kde=True, hue='ProductGroup', palette='coolwarm')
plt.title("Histogram with Top Products and 'Other'")
plt.show()
```

```python
#transaction
fig, axs=plt.subplots(1,1, figsize=(15,7))
plt.subplot(121)
sns.histplot(data=transaction,x='TotalValue',bins=10,kde=True,color='green')
plt.show()
```

```python
transaction['TransactionDate'] = pd.to_datetime(transaction['TransactionDate'])
transaction['TransactionMonth'] = transaction['TransactionDate'].dt.strftime('%Y-%m')

fig, axs=plt.subplots(1,1, figsize=(15,10))
sns.histplot(data=transaction,x='TransactionDate',kde=True,hue='Quantity')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Group by CustomerID and calculate the total value
top_customers = transaction.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(20)

top_customers = top_customers.reset_index()

# Plot the top 20 customers
plt.figure(figsize=(15, 8))
sns.barplot(data=top_customers, x='CustomerID', y='TotalValue', palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 20 Customers by Total Value', fontsize=16)
plt.xlabel('Customer ID', fontsize=14)
plt.ylabel('Total Value', fontsize=14)
plt.tight_layout()
plt.show()
```
